```python
ratings_name=ratings.merge(books,on="ISBN")
ratings_name.head(50)
```

```python
num_rating=ratings_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating.rename(columns={"Book-Rating":"No.of Ratings"},inplace=True)
num_rating.head(50)

```

```python
avg_rating=ratings_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating.rename(columns={"Book-Rating":"Avg rating"},inplace=True)
avg_rating.head(50)

```

```python

popular= num_rating.merge(avg_rating,on="Book-Title")

popular[popular["No.of Ratings"]>=250].sort_values("Avg rating",ascending=False)

popular.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','No.of Ratings','Avg rating']]


#sorting the popular datasets on the basis of Avg_rating
```

```python
x=ratings_name.groupby('User-ID').count()['Book-Rating']>200
learned_users= x[x].index

learned_users
#dataframe for learned users whose book rating is greater than 200

```

```python

x[x]
```

```python
filtered_rating=ratings_name[ratings_name['User-ID'].isin(learned_users)]
filtered_rating
```

```python
# now we are filtering the filtered rating on the basis of book title
y= filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50

famousbooks=y[y].index
famousbooks
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famousbooks)]
final_rating

```

```python
final_rating.drop_duplicates()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity

```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores.shape


```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    
    for i in similar_items:
            print(pt.index[i[0]])
    
```
